```python
"""
Optimized Rule-based / Heuristic solution for Lab 7 Part 1.

Strategy: High-speed Expectimax search with Precomputed Row Lookup Tables.
  - Precomputes all possible 2048 row transitions (16-bit encoding) for fast execution.
  - Eliminates heavy NumPy/rotation overhead during search.
  - Adaptive depth control (up to depth 7-8 in critical endgames).
  - Advanced Snake/Corner Matrix Heuristic with dynamic weight scaling.
"""

import argparse
import copy
import random
import numpy as np

# ---------------------------------------------------------------------------
# Constants
# ---------------------------------------------------------------------------
BOARD_SIZE = 4
TARGET_TILE = 2048

ACTION_UP = 0
ACTION_LEFT = 1
ACTION_DOWN = 2
ACTION_RIGHT = 3
N_ACTIONS = 4

# ---------------------------------------------------------------------------
# 1. Ultra-Fast Row Lookup Table Generation
# ---------------------------------------------------------------------------
_ROW_MOVE_TABLE = [0] * 65536
_ROW_SCORE_TABLE = [0] * 65536

def _precompute_tables():
    """Generates lookup tables for a single row sliding left."""
    for i in range(65536):
        t0 = (i >> 12) & 0xF
        t1 = (i >> 8) & 0xF
        t2 = (i >> 4) & 0xF
        t3 = i & 0xF

        row = [t0, t1, t2, t3]
        tiles = [v for v in row if v != 0]
        next_row = []
        score = 0
        skip = False

        for j in range(len(tiles)):
            if skip:
                skip = False
                continue
            if j + 1 < len(tiles) and tiles[j] == tiles[j + 1]:
                val = tiles[j] + 1
                next_row.append(val)
                score += (1 << val)
                skip = True
            else:
                next_row.append(tiles[j])

        while len(next_row) < 4:
            next_row.append(0)

        out_encoded = (next_row[0] << 12) | (next_row[1] << 8) | (next_row[2] << 4) | next_row[3]
        _ROW_MOVE_TABLE[i] = out_encoded
        _ROW_SCORE_TABLE[i] = score

# Initialize lookup tables immediately on import
_precompute_tables()

# ---------------------------------------------------------------------------
# High-Performance Board Representations
# ---------------------------------------------------------------------------
def _board_to_log_list(board):
    """Converts standard board to a flat 16-element log2 list."""
    out = [0] * 16
    flat = board.ravel()
    for i in range(16):
        v = flat[i]
        if v > 0:
            out[i] = int(v).bit_length() - 1
    return out

def _log_list_to_board(log_list):
    """Converts a flat log2 list back to a standard 4x4 NumPy array."""
    board = np.zeros((4, 4), dtype=np.int32)
    flat = board.ravel()
    for i in range(16):
        if log_list[i] > 0:
            flat[i] = 1 << log_list[i]
    return board

def _move_board(state, action):
    """Applies an action to a flat log2 state using precomputed tables."""
    next_state = list(state)
    score = 0

    if action == ACTION_LEFT:
        r0 = (state[0]<<12) | (state[1]<<8) | (state[2]<<4) | state[3]
        m0 = _ROW_MOVE_TABLE[r0]
        next_state[0], next_state[1], next_state[2], next_state[3] = (m0>>12)&0xF, (m0>>8)&0xF, (m0>>4)&0xF, m0&0xF
        score += _ROW_SCORE_TABLE[r0]

        r1 = (state[4]<<12) | (state[5]<<8) | (state[6]<<4) | state[7]
        m1 = _ROW_MOVE_TABLE[r1]
        next_state[4], next_state[5], next_state[6], next_state[7] = (m1>>12)&0xF, (m1>>8)&0xF, (m1>>4)&0xF, m1&0xF
        score += _ROW_SCORE_TABLE[r1]

        r2 = (state[8]<<12) | (state[9]<<8) | (state[10]<<4) | state[11]
        m2 = _ROW_MOVE_TABLE[r2]
        next_state[8], next_state[9], next_state[10], next_state[11] = (m2>>12)&0xF, (m2>>8)&0xF, (m2>>4)&0xF, m2&0xF
        score += _ROW_SCORE_TABLE[r2]

        r3 = (state[12]<<12) | (state[13]<<8) | (state[14]<<4) | state[15]
        m3 = _ROW_MOVE_TABLE[r3]
        next_state[12], next_state[13], next_state[14], next_state[15] = (m3>>12)&0xF, (m3>>8)&0xF, (m3>>4)&0xF, m3&0xF
        score += _ROW_SCORE_TABLE[r3]

    elif action == ACTION_RIGHT:
        r0 = (state[3]<<12) | (state[2]<<8) | (state[1]<<4) | state[0]
        m0 = _ROW_MOVE_TABLE[r0]
        next_state[3], next_state[2], next_state[1], next_state[0] = (m0>>12)&0xF, (m0>>8)&0xF, (m0>>4)&0xF, m0&0xF
        score += _ROW_SCORE_TABLE[r0]

        r1 = (state[7]<<12) | (state[6]<<8) | (state[5]<<4) | state[4]
        m1 = _ROW_MOVE_TABLE[r1]
        next_state[7], next_state[6], next_state[5], next_state[4] = (m1>>12)&0xF, (m1>>8)&0xF, (m1>>4)&0xF, m1&0xF
        score += _ROW_SCORE_TABLE[r1]

        r2 = (state[11]<<12) | (state[10]<<8) | (state[9]<<4) | state[8]
        m2 = _ROW_MOVE_TABLE[r2]
        next_state[11], next_state[10], next_state[9], next_state[8] = (m2>>12)&0xF, (m2>>8)&0xF, (m2>>4)&0xF, m2&0xF
        score += _ROW_SCORE_TABLE[r2]

        r3 = (state[15]<<12) | (state[14]<<8) | (state[13]<<4) | state[12]
        m3 = _ROW_MOVE_TABLE[r3]
        next_state[15], next_state[14], next_state[13], next_state[12] = (m3>>12)&0xF, (m3>>8)&0xF, (m3>>4)&0xF, m3&0xF
        score += _ROW_SCORE_TABLE[r3]

    elif action == ACTION_UP:
        c0 = (state[0]<<12) | (state[4]<<8) | (state[8]<<4) | state[12]
        m0 = _ROW_MOVE_TABLE[c0]
        next_state[0], next_state[4], next_state[8], next_state[12] = (m0>>12)&0xF, (m0>>8)&0xF, (m0>>4)&0xF, m0&0xF
        score += _ROW_SCORE_TABLE[c0]

        c1 = (state[1]<<12) | (state[5]<<8) | (state[9]<<4) | state[13]
        m1 = _ROW_MOVE_TABLE[c1]
        next_state[1], next_state[5], next_state[9], next_state[13] = (m1>>12)&0xF, (m1>>8)&0xF, (m1>>4)&0xF, m1&0xF
        score += _ROW_SCORE_TABLE[c1]

        c2 = (state[2]<<12) | (state[6]<<8) | (state[10]<<4) | state[14]
        m2 = _ROW_MOVE_TABLE[c2]
        next_state[2], next_state[6], next_state[10], next_state[14] = (m2>>12)&0xF, (m2>>8)&0xF, (m2>>4)&0xF, m2&0xF
        score += _ROW_SCORE_TABLE[c2]

        c3 = (state[3]<<12) | (state[7]<<8) | (state[11]<<4) | state[15]
        m3 = _ROW_MOVE_TABLE[c3]
        next_state[3], next_state[7], next_state[11], next_state[15] = (m3>>12)&0xF, (m3>>8)&0xF, (m3>>4)&0xF, m3&0xF
        score += _ROW_SCORE_TABLE[c3]

    elif action == ACTION_DOWN:
        c0 = (state[12]<<12) | (state[8]<<8) | (state[4]<<4) | state[0]
        m0 = _ROW_MOVE_TABLE[c0]
        next_state[12], next_state[8], next_state[4], next_state[0] = (m0>>12)&0xF, (m0>>8)&0xF, (m0>>4)&0xF, m0&0xF
        score += _ROW_SCORE_TABLE[c0]

        c1 = (state[13]<<12) | (state[9]<<8) | (state[5]<<4) | state[1]
        m1 = _ROW_MOVE_TABLE[c1]
        next_state[13], next_state[9], next_state[5], next_state[1] = (m1>>12)&0xF, (m1>>8)&0xF, (m1>>4)&0xF, m1&0xF
        score += _ROW_SCORE_TABLE[c1]

        c2 = (state[14]<<12) | (state[10]<<8) | (state[6]<<4) | state[2]
        m2 = _ROW_MOVE_TABLE[c2]
        next_state[14], next_state[10], next_state[6], next_state[2] = (m2>>12)&0xF, (m2>>8)&0xF, (m2>>4)&0xF, m2&0xF
        score += _ROW_SCORE_TABLE[c2]

        c3 = (state[15]<<12) | (state[11]<<8) | (state[7]<<4) | state[3]
        m3 = _ROW_MOVE_TABLE[c3]
        next_state[15], next_state[11], next_state[7], next_state[3] = (m3>>12)&0xF, (m3>>8)&0xF, (m3>>4)&0xF, m3&0xF
        score += _ROW_SCORE_TABLE[c3]

    changed = (state != next_state)
    return next_state, score, changed

# ---------------------------------------------------------------------------
# Snake/Corner Matrix Heuristic Function
# ---------------------------------------------------------------------------
SNAKE_MATRIX = [
    15, 14, 13, 12,
     8,  9, 10, 11,
     7,  6,  5,  4,
     0,  1,  2,  3
]

def _evaluate_board(state):
    empty_count = 0
    snake_score = 0.0
    smoothness_penalty = 0.0

    for i in range(16):
        val = state[i]
        if val == 0:
            empty_count += 1
            continue

        snake_score += (1 << val) * SNAKE_MATRIX[i]

        row, col = i // 4, i % 4
        if col < 3:
            r_val = state[i + 1]
            if r_val > 0:
                smoothness_penalty -= abs(val - r_val) * 12.0
        if row < 3:
            d_val = state[i + 4]
            if d_val > 0:
                smoothness_penalty -= abs(val - d_val) * 12.0

    if empty_count > 0:
        if empty_count <= 4:
            empty_score = (1 << 14) * empty_count
        else:
            empty_score = (1 << 11) * empty_count
    else:
        empty_score = -5000.0

    return snake_score + smoothness_penalty + empty_score

# ---------------------------------------------------------------------------
# Expectimax with Adaptive Pruning & Deepening
# ---------------------------------------------------------------------------
def _expectimax(state, depth, is_player_turn, spawn_two_prob=0.9):
    if depth == 0:
        return _evaluate_board(state)

    if is_player_turn:
        best = -float('inf')
        any_valid = False
        for action in range(4):
            next_s, _, changed = _move_board(state, action)
            if not changed:
                continue
            any_valid = True
            val = _expectimax(next_s, depth - 1, False, spawn_two_prob)
            if val > best:
                best = val
        return best if any_valid else _evaluate_board(state)

    else:
        empty_indices = [i for i in range(16) if state[i] == 0]
        if not empty_indices:
            return _evaluate_board(state)

        if len(empty_indices) > 5:
            empty_indices = empty_indices[:5]

        total_val = 0.0
        n = len(empty_indices)

        for idx in empty_indices:
            state[idx] = 1
            v2 = _expectimax(state, depth - 1, True, spawn_two_prob)

            state[idx] = 2
            v4 = _expectimax(state, depth - 1, True, spawn_two_prob)

            state[idx] = 0
            total_val += (v2 * spawn_two_prob + v4 * (1.0 - spawn_two_prob)) / n

        return total_val

def _choose_best_action(state, spawn_two_prob=0.9):
    empty_count = state.count(0)

    if empty_count >= 10:
        depth = 3
    elif empty_count >= 6:
        depth = 4
    elif empty_count >= 4:
        depth = 5
    else:
        depth = 7

    best_action = ACTION_LEFT
    best_score = -float('inf')

    for action in [ACTION_LEFT, ACTION_UP, ACTION_DOWN, ACTION_RIGHT]:
        next_s, _, changed = _move_board(state, action)
        if not changed:
            continue
        score = _expectimax(next_s, depth - 1, False, spawn_two_prob)
        if score > best_score:
            best_score = score
            best_action = action

    return best_action

# ---------------------------------------------------------------------------
# Required Interfaces
# ---------------------------------------------------------------------------
class Game2048Env:
    def __init__(self, seed=0, spawn_two_prob=0.90):
        self.rng = random.Random(seed)
        self.spawn_two_prob = float(spawn_two_prob)
        self.board = np.zeros((BOARD_SIZE, BOARD_SIZE), dtype=np.int32)
        self.score = 0
        self.steps = 0
        self.reset(seed=seed)

    def reset(self, seed=None, options=None):
        if seed is not None:
            self.rng.seed(seed)
        self.board = np.zeros((BOARD_SIZE, BOARD_SIZE), dtype=np.int32)
        self.score = 0
        self.steps = 0
        self.add_random_tile(self.spawn_two_prob)
        self.add_random_tile(self.spawn_two_prob)
        return self.state(), self.info()

    def step(self, action):
        action = int(np.asarray(action).item())
        self.steps += 1

        current_state_list = _board_to_log_list(self.board)
        next_state_list, score_gained, board_changed = _move_board(current_state_list, action)

        if board_changed:
            self.board = _log_list_to_board(next_state_list)
            self.score += score_gained
            self.add_random_tile(self.spawn_two_prob)

        terminated = not self.any_valid_action()
        truncated = False
        return self.state(), float(score_gained), terminated, truncated, self.info()

    def add_random_tile(self, spawn_two_prob):
        empty = list(zip(*np.where(self.board == 0)))
        if not empty:
            return False
        y, x = self.rng.choice(empty)
        value = 2 if self.rng.random() < spawn_two_prob else 4
        self.board[y, x] = value
        return True

    def valid_actions(self):
        state_list = _board_to_log_list(self.board)
        return [
            _move_board(state_list, ACTION_UP)[2],
            _move_board(state_list, ACTION_LEFT)[2],
            _move_board(state_list, ACTION_DOWN)[2],
            _move_board(state_list, ACTION_RIGHT)[2]
        ]

    def any_valid_action(self):
        return any(self.valid_actions())

    def state(self):
        obs = np.zeros_like(self.board, dtype=np.float32)
        nonzero = self.board > 0
        obs[nonzero] = np.log2(self.board[nonzero]) / 16.0
        return obs.reshape(-1)

    def info(self):
        max_tile = int(np.max(self.board)) if self.board.size > 0 else 0
        success = bool(max_tile >= TARGET_TILE)
        return {
            "score": int(self.score),
            "max_tile": max_tile,
            "steps": int(self.steps),
            "success": success,
            "is_success": success,
        }


class RuleBasedPolicy:
    def __init__(self, env=None):
        self.env = env

    def predict(self, obs, deterministic=True):
        state_list = _board_to_log_list(self.env.board)
        action = _choose_best_action(state_list, spawn_two_prob=self.env.spawn_two_prob)
        return int(action), None


class Game2048Agent:
    def __init__(self, seed=0):
        self.seed = seed
        self.env = Game2048Env(seed=seed)
        self.model = None

    def load(self):
        self.model = RuleBasedPolicy(env=self.env)
        return self.model

    def train(self, timesteps=0):
        self.load()
        return self.model

    def test(self, games=5, random_policy=False):
        if isinstance(games, int):
            test_seeds = [self.seed + 1000 + game for game in range(games)]
        else:
            test_seeds = list(games)

        if self.model is None:
            self.load()

        for game, seed in enumerate(test_seeds):
            obs, _ = self.env.reset(seed=seed)
            done = False
            info = self.env.info()

            while not done:
                if random_policy:
                    action = self.env.rng.randrange(N_ACTIONS)
                else:
                    action, _ = self.model.predict(obs, deterministic=True)
                obs, _, terminated, truncated, info = self.env.step(action)
                done = terminated or truncated

            print(
                f"game={game + 1} seed={seed} score={info['score']} "
                f"max_tile={info['max_tile']} steps={info['steps']}"
            )


def main():
    parser = argparse.ArgumentParser()
    parser.add_argument("--test-games", type=int, default=5)
    parser.add_argument("--test-seeds", type=int, nargs="*", default=None)
    parser.add_argument("--seed", type=int, default=0)
    parser.add_argument("--random", action="store_true")
    args = parser.parse_args()

    agent = Game2048Agent(seed=args.seed)
    agent.load()
    test_games = args.test_seeds if args.test_seeds is not None else args.test_games
    agent.test(games=test_games, random_policy=args.random)


if __name__ == "__main__":
    main()
```
